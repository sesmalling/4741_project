```python
import os
import pandas as pd
import numpy as np
import re 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from typing import List, Dict
```

```python
pd.set_option("display.max_columns", 500)
```

```python
years = range(2016, 2020)
dataset_filenames = [os.path.join("data", f"clean_{year}.csv") for year in years]

d16, d17, d18, d19 = (pd.read_csv(filename) for filename in dataset_filenames)

d17_19 = pd.read_csv(os.path.join("data", "clean_2017-2019.csv"))
```

```python
data_diagnosed = d17_19[d17_19.ever_diagnosed.notnull() & data_diagnosed.age.notnull()]
```

```python
# data_diagnosed.ever_diagnosed.unique()
data_diagnosed.loc[data_diagnosed.ever_diagnosed == 'Yes', 'ever_diagnosed'] = 1
data_diagnosed.loc[data_diagnosed.ever_diagnosed == 'No', 'ever_diagnosed'] = 0
# print(data_diagnosed.ever_diagnosed.unique())
data_diagnosed.head()
```

```python
def mse(y: np.ndarray, pred: np.ndarray) -> float:
    """
    Returns: the mean squared error between y and pred. 
    """
    return ((y - pred) ** 2).sum() / y.shape[0]


def plot_data(x: np.ndarray, y: np.ndarray, xname: str, yname: str, margin:float=0.05) -> None:
    """
    Plots the points x,y on a scatterplot with xlabel xname and ylabel yname. 
    """
    data = plt.scatter(x, y, label='Data', c='magenta')
    plt.xlabel(xname)
    plt.ylabel(yname)
    range_y = y.max() - y.min()
    range_x = x.max() - x.min()
    xsamples = [
        x.min() - margin * range_x, x.max() + margin * range_x
    ]
    plt.plot(xsamples, xsamples, c='b')
    plt.ylim((0.0, y.max() + margin * range_y))
    plt.xlim((0.0, x.max() + margin * range_x))
    
    # line_up, = plt.plot([1, 2, 3], label='Line 2')
    # line_down, = plt.plot([3, 2, 1], label='Line 1')
    # plt.legend([line_up, line_down], ['Line Up', 'Line Down'])
    # plt.scatter(x, y, label=)

def plot_pred_true(test_pred, test_y):
    plot_data(test_pred, test_y, "Predicted Ever Diagnosed", "Actual Ever Diagnosed")
    
def get_incorrectly_classified_numbers(y: np.ndarray, pred: np.ndarray):
    adjusted_pred = np.where(pred < 0.5, 0, 1)
    incorrectly_classified = np.logical_xor(y, adjusted_pred).sum()
    return incorrectly_classified
```

```python
def hcat(matrices=[], arrays=[]):
    return np.hstack([
        *map(np.array, matrices),
        *(np.array([arr]).T for arr in arrays)
    ])
```

```python
train_proportion = 0.8

was_diagnosed = data_diagnosed.ever_diagnosed.to_numpy()
data = data_diagnosed.age.to_numpy()
num_entries = len(data)
X_train, X_test, y_train, y_test = train_test_split(data, was_diagnosed, train_size=train_proportion)
num_train = len(X_train)
X_train = np.vstack([X_train, np.ones(num_train)]).T
X_test = np.vstack([X_test, np.ones(num_entries - num_train)]).T
```

```python
w = np.array(np.linalg.lstsq(X_train, y_train, rcond=None)[0])
train_pred = X_train @ w
test_pred = X_test @ w

train_mse = mse(y_train, train_pred) 
test_mse = mse(y_test, test_pred)
print("Train MSE\t", train_mse)
print("Test MSE \t", test_mse)
print(get_incorrectly_classified_numbers(test_pred, y_test))
plot_pred_true(test_pred, y_test)
```

```python
def remove_punctuation(s: str) -> str:
    """
    Removes punctuation from string s. 
    """
    return re.sub(r'[^\w\s]', '', s)

def remove_whitespace(s: str) -> str:
    """
    Removes whitespace from string s.
    """
    return re.sub(r'[ \t\r\n]', '', s)

def normalize(s: str) -> str:
    """
    Normalizes string s to remove punctuation and whitespace, and lowercased.
    """
    return remove_whitespace(remove_punctuation(s)).lower()
```

```python
genders_in_dataset = list(data_diagnosed.gender.unique())

MALE_TERMS = ['m', 'male', 'man']
FEMALE_TERMS = ['f', 'female', 'woman', 'femail'] # handle misspellings
QUEER_TERMS = ['nonconforming', 'queer', 'fluid', 'nonbinary', 'trans', 'agender', 'questioning', 'nb']

def get_normalized_gender(gender: str) -> str:
    if type(gender) != str and np.isnan(gender):
        return 'NaN'
    gender_depunct = normalize(gender)
    if any(term in gender_depunct for term in QUEER_TERMS):
        return 'Nonbinary'
    if any(term in gender_depunct for term in FEMALE_TERMS):
        return 'Female'
    if any(term in gender_depunct for term in MALE_TERMS):
        return 'Male'
    return 'Nonbinary'

gender_mappings = { gender: get_normalized_gender(gender) for gender in genders_in_dataset }
```

```python
was_diagnosed = data_diagnosed.ever_diagnosed.to_numpy()
data = data_diagnosed[['age', 'gender']].to_numpy()
gender_from_dataset = data[:, 1]
mapped_genders = [get_normalized_gender(gender) for gender in gender_from_dataset]
```

```python
#Sets of all categories in a particular column
gender_labels = set(gender_mappings.values())

def one_hot_matrix(column_of_entries: List[str], possible_entries_in_column: List[str]) -> List[List[int]]:
    """
    Returns: a one-hot encoding of every possible value in column_of_entries, as a 2d list. 
    """
    return many_hot_matrix([[entry] for entry in column_of_entries], possible_entries_in_column)

def make_many_hot_vector(values_in_set: List[str], indices_dict: Dict[str, int], num_entries: int) -> List[int]:
    """
    Returns: the many hot vector corresponding to the values_in_set, mapped with indices_dict.
    """
    vec = [0] * num_entries
    for value in values_in_set:
        vec[indices_dict[value]] = 1
    return vec

def many_hot_matrix(column_of_entries: List[List[str]], possible_entries_in_column: List[str]) -> List[List[int]]:
    """
    Returns: a many-hot encoding of every possible value in column_of_entries, as a 2d list.
    """
    indices_dict = {element : i for i,element in enumerate(possible_entries_in_column)}
    num_entries = len(indices_dict)
    return [make_many_hot_vector(entry, indices_dict, num_entries) for entry in column_of_entries]
```

```python
data = hcat(
    matrices=[one_hot_matrix(mapped_genders, gender_labels)],
    arrays=[data_diagnosed.age]
)
num_entries = len(data)

X_train, X_test, y_train, y_test = train_test_split(data, was_diagnosed, train_size=train_proportion)

num_train = len(X_train)

X_train = hcat(
    matrices=[X_train],
    arrays=[np.ones(num_train)]
)

X_test = hcat(
    matrices = [X_test],
    arrays = [np.ones(num_entries - num_train)]
)
```

```python
w_age_gender = np.array(np.linalg.lstsq(X_train, y_train, rcond=None)[0])
train_pred = X_train @ w_age_gender
test_pred = X_test @ w_age_gender

train_mse = mse(y_train, train_pred) 
test_mse = mse(y_test, test_pred)
print("Train MSE\t", train_mse)
print("Test MSE \t", test_mse)
print(get_incorrectly_classified_numbers(test_pred, y_test))
plot_pred_true(test_pred, y_test)
```
